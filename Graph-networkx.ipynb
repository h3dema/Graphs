# Este notebook utiliza uma biblioteca do python para DFS e BFS

Recomendo utilizá-la ao invés de implementações manuais

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
G = nx.Graph()
```

```python
G.add_edge('a', 'b')
G.add_edge('a', 's')
G.add_edge('s', 'b')
G.add_edge('s', 'c')
G.add_edge('s', 'e')
G.add_edge('f', 'e')
G.add_edge('f', 'c')
G.add_edge('f', 'd')
G.add_edge('e', 'g')
G.add_edge('e', 'd')
G.add_edge('d', 'g')
G.add_edge('d', 'c')
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
nx.draw_networkx(G, arrows=False, with_labels=True, font_size=20)
```

# DFS - depth-first-search

```python
dfs = nx.dfs_tree(G, 's')
```

```python
nx.draw_networkx(dfs, arrows=False, with_labels=True)
```

# BFS breadth-first-search

```python
bfs = nx.bfs_tree(G, 's')
```

```python
nx.draw_networkx(bfs, arrows=False, with_labels=True)
```
