# Este notebook utiliza uma biblioteca do python para DFS e BFS

Recomendo utilizá-la ao invés de implementações manuais

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
G = nx.DiGraph()
```

![Direct graph](digrafo.png)

```python
G.add_edge('s', 'a')
G.add_edge('a', 'b')
G.add_edge('b', 's')
G.add_edge('s', 'c')
G.add_edge('s', 'e')
G.add_edge('c', 'd')
G.add_edge('d', 'g')
G.add_edge('g', 'e')
G.add_edge('e', 'd')
G.add_edge('e', 'f')
G.add_edge('f', 'd')
G.add_edge('f', 'c')
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
nx.draw_networkx(G, with_labels=True, font_size=20)
```

# DFS - depth-first-search

```python
dfs = nx.dfs_tree(G, 's')
```

```python
nx.draw_networkx(dfs, with_labels=True)
```

# BFS breadth-first-search

```python
bfs = nx.bfs_tree(G, 's')
```

```python
nx.draw_networkx(bfs, with_labels=True)
```

## Strong Connect Components

```python
cc = nx.strongly_connected_components(G)
```

```python
# tamanho dos componentes conectados
[len(c) for c in sorted(cc, key=len, reverse=True)]
```
